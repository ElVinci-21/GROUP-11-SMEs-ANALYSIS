```python
import statsmodels.api as sm

# Select variables and remove rows with missing values
regression_data = df[[
    "log profitability",
    "log loan amount",
    "log firm age",
    "tax rate",
    "log employees"
]].dropna()

# Dependent variable
y = regression_data["log profitability"]

# Independent variables
X = regression_data[[
    "log loan amount",
    "log firm age",
    "tax rate",
    "log employees"
]]

# Add intercept (constant)
X = sm.add_constant(X)

# Fit OLS model
model = sm.OLS(y, X).fit()

# Display results
print(model.summary())
```

```python
# Predicted values
regression_data["predicted_log_profitability"] = model.predict(X)

# Residuals
regression_data["residuals"] = model.resid
```

```python
n = len(regression_data)               # sample size
k = len(X.columns) - 1    # number of predictors

SSR = model.ess           # Regression Sum of Squares
SSE = model.ssr           # Error Sum of Squares
SST = SSR + SSE

regression_data_reg = k
regression_data_err = n - k - 1
regression_data_tot = n - 1

MSR = SSR / regression_data_reg
MSE = SSE / regression_data_err

F = MSR / MSE

anova_table = pd.DataFrame({
    'Source':['Regression','Residual','Total'],
    'SS':[SSR,SSE,SST],
    'df':[regression_data_reg,regression_data_err,regression_data_tot],
    'MS':[MSR,MSE,''],
    'F':[F,'','']
})

print(anova_table)
```
